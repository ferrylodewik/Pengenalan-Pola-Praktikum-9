```python
from matplotlib import pyplot
from numpy.random import normal
```

```python
sample = normal (size=1000)

pyplot.hist(sample, bins=20)
pyplot.show()
```

```python
from numpy import mean
from numpy import std
from scipy.stats import norm
```

```python
sample = normal (loc=50, scale=5, size=1000)
```

```python
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean = %.3f, Standar & Standard Deviation = %.3f' % (sample_mean, sample_std))
```

```python
dist = norm(sample_mean, sample_std)
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]
```

```python
pyplot.hist(sample, bins=10, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

```python
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity
```

```python
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
```

```python
pyplot.hist(sample, bins=20)
pyplot.show()
```

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
```

```python
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
```

```python
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```

Lakukan beberapa modifikasi pada nilai bin dan sampel serta visualisasikan histogram tersebut dan sajikan pada box di bawah ini.
Sample : 1000, 3000, 5000 & bin : 20, 40, 60

```python
from matplotlib import pyplot
from numpy.random import normal
```

```python
sampel = [1000, 3000, 5000]
bin = [20, 40, 60]
for i in range(len(sampel)):
  for j in range(len(bin)):
    sampel_normal = normal(size=sampel[i])
    pyplot.hist(sampel_normal, bins=bin[j])
    pyplot.title('Untuk Sampel = ' + str(sampel[i]) + ' dan bin = ' + str(bin[j]))
    pyplot.show()
```

No. 2
Pada estimasi kepekatan parametrik, cari distribusi lain selain distribusi normal dan lakukan visualiasi

```python
from matplotlib import pyplot
from scipy.stats import gamma
from scipy.stats import expon
import numpy as np
```

Distribusi Gamma

```python
sampel_gamma = gamma.rvs(a=2, size=1000)
x2 = np.linspace(0, 10, 1000)
probabilities = gamma.pdf(x2, a=2)
pyplot.hist(sampel_gamma, bins=20, density=True)
pyplot.plot(x2, probabilities)
pyplot.title('Distribusi Gamma')
pyplot.show()
```

Distribusi Eksponensial

```python
sampel_exp = expon.rvs(scale=5, size=1000)
x1 = np.linspace(0, 35, 1000)
probabilities = expon.pdf(x1, scale=5)
pyplot.hist(sampel_exp, bins=20, density=True)
pyplot.plot(x1, probabilities)
pyplot.title('Distribusi Eksponensial')
pyplot.show()
```

No. 3
Buat histogram untuk contoh pembangkitan data dengan jumlah peak = 3 (multimodal) kemudian gunakan fungsi kernel density estimation dan lakukan visualisasi

```python
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
```

```python
sampel1 = normal(loc=20, scale=5, size=300)
sampel2 = normal(loc=40, scale=5, size=500)
sampel3 = normal(loc=60, scale=5, size=700)
sampel = hstack((sampel1, sampel2, sampel3))
```

```python
pyplot.hist(sampel, bins=50)
pyplot.show()
model = KernelDensity(bandwidth=3, kernel='gaussian')
sampel = sampel.reshape((len(sampel), 1))
model.fit(sampel)
```

```python
values = asarray([value for value in range(1, 80)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
pyplot.hist(sampel, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```
